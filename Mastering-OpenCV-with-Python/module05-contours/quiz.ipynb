```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python

img = cv2.imread('Quiz5_1.jpg', cv2.IMREAD_COLOR)
# error findContours applied only for gray images
#contours, hierarchy = cv2.findContours( img, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
#print('Number of contours : ', len(contours))
plt.imshow(img[...,::-1]);
```

```python
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ret1, thresh1 = cv2.threshold(imgGray, 127, 255, cv2.THRESH_BINARY)
contours1, hierarchy1 = cv2.findContours( thresh1, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

ret2, thresh2 = cv2.threshold(imgGray, 127, 255, cv2.THRESH_BINARY_INV) # inversed
contours2, hierarchy2 = cv2.findContours( thresh2, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

print('Number of contours case 1 : ', len(contours1))
print('Number of contours case 2 : ', len(contours2))

plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(thresh1, cmap='gray')
plt.title('THRESH_BINARY')

plt.subplot(1, 2, 2)
plt.imshow(thresh2, cmap='gray')
plt.title('THRESH_BINARY_INV')

plt.tight_layout()
```

```python
# Create a copy of the original image to draw contours on
img_with_contours = img.copy()

# Draw all contours on the image
# -1 means draw all contours, (0, 255, 0) is green color, 2 is the thickness
cv2.drawContours(img_with_contours, contours1, -1, (0, 255, 0), 2)

# Display the image with contours
plt.figure(figsize=(10, 8))
plt.imshow(img_with_contours[...,::-1])  # Convert BGR to RGB for matplotlib
plt.title('Image with Contours from THRESH_BINARY')
plt.axis('off')
```

```python
img_5_3 = cv2.imread('Quiz5_3_2_v2.png', cv2.IMREAD_COLOR)

# Add border
# Parameters: src, top, bottom, left, right, borderType, value
bordered_img = cv2.copyMakeBorder(img_5_3, 20, 20, 20, 20, cv2.BORDER_CONSTANT, value=[255,0,0]) # blue

# Apply Gaussian blur to img_5_3
# Parameters: src, ksize (width, height), sigmaX
blurred_img = cv2.GaussianBlur(img_5_3, (5, 5), 0)

# Apply Erosion
# Create a kernel for erosion (structuring element)
kernel = np.ones((8, 8), np.uint8)

# Apply erosion to img_5_3
eroded_img = cv2.erode(img_5_3, kernel, iterations=1)

# Display results
plt.figure(figsize=(15, 8))

plt.subplot(1, 4, 1)
plt.imshow(img_5_3[...,::-1])
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(bordered_img[:,:,::-1])
plt.title('Image with Border')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(blurred_img[...,::-1])  # Convert BGR to RGB for matplotlib
plt.title('Gaussian Blurred Image (5x5 kernel)')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(eroded_img[...,::-1])  # Convert BGR to RGB for matplotlib
plt.title('Eroded Image (5x5 kernel)')
plt.axis('off')

plt.tight_layout()
```
